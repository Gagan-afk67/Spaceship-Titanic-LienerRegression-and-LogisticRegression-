```python
import numpy as np 
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
train=pd.read_csv("spaceship-titanic/train.csv")
```

```python
print(train["CryoSleep"].mode())
train["CryoSleep"]=train["CryoSleep"].fillna("False")
```

```python
print(train["HomePlanet"].mode())
train["HomePlanet"]=train["HomePlanet"].fillna("Earth")
```

```python
print(train["Cabin"].mode())
train["Cabin"]=train["Cabin"].fillna("G/734/S")
```

```python
print(train["Destination"].mode())
train["Destination"]=train["Destination"].fillna("TRAPPIST-1e")
```

```python
print(train["VIP"].mode())
train["VIP"]=train["VIP"].fillna("False")
```

```python
# Numerical columns
# filled every null value of num or numerical columns at once 
# the fillna will find the mean of every train num columns 
num_cols = ['Age', 'RoomService', 'FoodCourt', 'ShoppingMall', 'Spa', 'VRDeck']
train[num_cols] = train[num_cols].fillna(train[num_cols].mean())
```

```python
# to fill nan values of name 
train['Name']=train['Name'].fillna('Unknown')
```

```python
# ENCORDING converying cat to num 
# We encode categorical text data into numbers because most machine learning algorithms can only work with numbers, not text.
from sklearn.preprocessing import LabelEncoder
LE=LabelEncoder()
```

```python
train.columns
```

```python
for i in train.columns:
    if train[i].dtype=="str":
        train[i]=LE.fit_transform(train[i])
    else:
        print(i)
```
